```python
import xlsxwriter
import os
from datetime import date
import shutil

class File:
  def __init__(self,file):
    self.filename = file.split(".")[0]
    self.format = file.split(".")[1]
    self.productKey = self.filename.split("_")[0]
    self.sizeWithAmount = self.filename.split("_")[1]
    self.size = []
    self.color = self.filename.split("_")[2]
    self.priceNoDiscount = self.filename.split("_")[3].split(",")[0]
    self.priceDiscount = self.filename.split("_")[3].split(",")[1] if (self.filename.split("_")[3].split(",")[1] != self.priceNoDiscount) else ""
    self.description = self.filename.split("_")[4] if len(self.filename.split("_")) > 4 else ""
    self.descriptionString = self.description.replace("-"," ")
    if(len(self.sizeWithAmount.split(","))>1):
        for x in range(len(self.sizeWithAmount.split(","))):
            self.size.append(self.sizeWithAmount.split(",")[x].split("-")[0])
        self.size = ','.join(self.size)
    else:
        self.size = self.sizeWithAmount.split("-")[0]
    self.replacementname = self.productKey+"productimg"+"."+self.format
```

Function for Variations

```python
def addVariations(variationSizeQuantity,file):
    variationSize = variationSizeQuantity.split("-")[0]
    variationQuantity = variationSizeQuantity.split("-")[1]
    return {
            'typos':"variation",
            'kodikos': "",
            'onoma': "",
            'dimosieumeno':"1",
            'isProteinomeno':"0",
            'isOrato':"visible",
            'shortPerigrafi':"" ,
            'perigrafi':"" ,
            'startDate':"",
            'endDate':"",
            'statusTax':"taxable",
            'categoryTax':"parent" ,
            'seApothema':"1",
            'apothema':variationQuantity,
            'lowApothema':"",
            'allowWaitingList':"0",
            'singleSale':"0",
            'allowGrade':"0",
            'noteOfSale':"",
            'discountPrice':file.priceDiscount,
            'price': file.priceNoDiscount,
            'categories': "" ,
            'etiketes':"",
            'class':"",
            'images': "" ,
            'limitMetafortiseis':"",
            'dateMetafortiseis':"",
            'parent': file.productKey,
            'productGroup':"",
            'saleRaise':"",
            'propertyGroup1': "Μέγεθος",
            'propertyGroup1Value': variationSize,
            'propertyGroup1Visibility':"1",
            'propertyGroup1Katholiki':"1",
            'propertyGroup2':"",
            'propertyGroup2Value': "",
            'propertyGroup2Visibility':"",
            'propertyGroup2Katholiki':"",
            'propertyGroup1Pre':"",
            'propertyGroup2Pre':""
            }
```
